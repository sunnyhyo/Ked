```python
import pandas as pd
kstat = pd.read_csv('./Main table HSCODE.csv')
```

```python
kstat.columns
```

```python
del kstat['일본품목명_2012']
del kstat['중국품목명_2012']

kstat = kstat.loc[kstat['한국품목명_2012'].notnull()].set_index(['HS_CODE_6', '한국품목명_2012'])
kstat
```

```python
kstat = kstat.reset_index(drop= True)
kstat
```

```python
stub = ['한국수출금액', '한국수출증감률', '한국수입금액', '한국수입증감률', '한국수지',
       '중국수출금액', '중국수출증감률', '중국수입금액', '중국수입증감률', '중국수지',
       '일본수출금액', '일본수출증감률', '일본수입금액', '일본수입증감률', '일본수지']
kstat2= pd.wide_to_long(kstat, 
                stubnames = stub, 
                i = ['HS_CODE_6', '한국품목명_2012'],
                j = '년도',
                sep= '_')#.reset_index()
```

```python
kstat2.columns
```

```python
a = ['수출금액_한국', '수출증감률_한국', '수입금액_한국', '수입증감률_한국', '수지_한국', 
 '수출금액_중국', '수출증감률_중국', '수입금액_중국', '수입증감률_중국', '수지_중국', 
 '수출금액_일본', '수출증감률_일본', '수입금액_일본', '수입증감률_일본', '수지_일본']
```

```python
kstat2.columns = a
kstat2.columns
```

```python
kstat2 = kstat2.reset_index()
kstat2
```

```python
kstat2.columns 
```

```python
stub = ['수출금액', '수출증감률', '수입금액', '수입증감률', '수지']
kstat3 = pd.wide_to_long(kstat2, 
                stubnames = stub, 
                i = ['HS_CODE_6', '한국품목명_2012', '년도'],
                j = '국가',
                sep= '_',
               suffix='.+').reset_index()
```
